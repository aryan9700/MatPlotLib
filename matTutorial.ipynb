```python
#Introduction to Matplotlib.pyplot
import matplotlib.pyplot as plt 
#running this import is compulsary to run below  different codes without import

#matplotlib supports the following graph
# line, scatter, bargraph, histogram, pichart
# we will plot a line from (1,3) and (4,8)
x=[1,2,3,4]
y=[1,4,9,16]
plt.title("Numbers and their squares")
#plt.xlabel("Numbers")
#plt.ylabel("Squares of given no.")
plt.plot(x,y, label='squares in y-axis', marker='d',mfc='w',mec='r',lw=5,c='k', ls='dotted') #to draw the line chart #or plt.plot(x,y,y,x) #lw= line width #c=color k for black #ls=line style=dashed or '--'
plt.plot(y,x, label='squares in x-axis',marker='*',mfc='r',mec='b',ms=10) #marker=d,o,*,+ #mfc=marcer face color=w,r,g,b,... #mec=marker edge color #ms=marker size to increase marker size
plt.legend()
plt.grid()
plt.show() #important to visible 
```

```python
#scatter plot
x=[1,2,3,4]
y=[1,4,9,16]
col=['r','b','g','y'] #color
font1={'family':'serif','color':'black','size':20}
plt.title("Numbers and their squares", fontdict=font1,loc='right') #loc=location 
#plt.xlabel("Numbers", fontdict=font1)
#plt.ylabel("Squares of given no.", fontdict=font1)
plt.scatter(x,y, label='squares in y-axis', c=col) #call c=col
plt.legend()
#plt.grid(axis='x', ls='--', c='b', lw=3)
plt.show() #important to visible 
```

```python
#bargraph
x=['python', 'c++' , 'java', 'HTML']
y=[1,5,8,2]
col=['red','blue','green','pink']
plt.barh(x,y,color=col) #bar 
plt.title("languages and their complexity")

#plt.xlabel("complexity")
#plt.ylabel("lang")
plt.show()
```

```python
#histogram =intervals, frequencies
import numpy as np

x=np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
#piechart
x=[4,6,2,5]
l=['python','c++','java','html']
col=['red','blue','green','yellow']
plt.pie(x,labels=l, startangle=90,explode=[0.02,0.05,0.07,0.01],shadow=True,colors=col)
plt.legend(title='languages')
plt.show()
```
